```python
# Analyze multiple products to understand format variations
print("=== Analyzing Product Format Variations ===")

# Get first page of products
response_data = response.json()
items = response_data.get('items', [])

print(f"Analyzing {min(len(items), 15)} products for format variations...\n")

price_formats = {
    'regular_price': 0,
    'lidl_plus_only': 0,
    'both_price_formats': 0,
    'no_price': 0,
    'discount_available': 0
}

for i, item in enumerate(items[:15]):
    data_dict = item.get('gridbox', {}).get('data', {})
    name = data_dict.get('fullTitle', 'No Name')
    product_id = data_dict.get('productId', 'No Product ID')
    
    print(f"--- Product {i+1}: {name[:50]}... ---")
    print(f"Product ID: {product_id}")
    
    # Check regular price structure
    regular_price_dict = data_dict.get('price', {})
    regular_price = regular_price_dict.get('price', None)
    old_price = regular_price_dict.get('oldPrice', None)
    
    # Check lidlPlus structure
    lidl_plus_array = data_dict.get('lidlPlus', [])
    lidl_plus_price = None
    if lidl_plus_array and len(lidl_plus_array) > 0:
        lidl_plus_price_dict = lidl_plus_array[0].get('price', {})
        lidl_plus_price = lidl_plus_price_dict.get('price', None)
    
    # Analyze packaging/unit info
    packaging_regular = regular_price_dict.get('packaging', {})
    packaging_text = packaging_regular.get('text', 'No unit info')
    
    # Check for discounts
    discount_dict = regular_price_dict.get('discount', {})
    has_discount = bool(discount_dict)
    
    # Categorize price format
    if regular_price and lidl_plus_price:
        price_formats['both_price_formats'] += 1
        format_type = "BOTH (Regular + LidlPlus)"
        final_price = regular_price
    elif regular_price and not lidl_plus_price:
        price_formats['regular_price'] += 1
        format_type = "REGULAR ONLY"
        final_price = regular_price
    elif not regular_price and lidl_plus_price:
        price_formats['lidl_plus_only'] += 1
        format_type = "LIDL PLUS ONLY"
        final_price = lidl_plus_price
    else:
        price_formats['no_price'] += 1
        format_type = "NO PRICE"
        final_price = None
    
    if has_discount:
        price_formats['discount_available'] += 1
    
    print(f"Format Type: {format_type}")
    print(f"Regular Price: €{regular_price if regular_price else 'None'}")
    print(f"LidlPlus Price: €{lidl_plus_price if lidl_plus_price else 'None'}")
    print(f"Old Price: €{old_price if old_price else 'None'}")
    print(f"Final Price: €{final_price if final_price else 'None'}")
    print(f"Unit/Packaging: {packaging_text}")
    print(f"Has Discount: {has_discount}")
    
    if has_discount:
        discount_text = discount_dict.get('discountText', 'No text')
        print(f"Discount Info: {discount_text}")
    
    # Check category
    category = data_dict.get('category', 'No category')
    print(f"Category: {category}")
    
    print("-" * 60)

print("\n=== Format Summary ===")
total_analyzed = sum(price_formats.values()) - price_formats['discount_available']  # discount is not mutually exclusive
for format_type, count in price_formats.items():
    if format_type != 'discount_available':
        percentage = (count / total_analyzed) * 100 if total_analyzed > 0 else 0
        print(f"{format_type.replace('_', ' ').title()}: {count} products ({percentage:.1f}%)")

print(f"Products with discounts: {price_formats['discount_available']} products")

print("\n=== Recommendations for Scraper ===")
print("1. Always check both 'price' and 'lidlPlus' structures")
print("2. Handle cases where only lidlPlus pricing is available")
print("3. Check for discount information in price.discount")
print("4. Extract unit information from price.packaging.text")
print("5. Use oldPrice for original price when available")
```
