```python
# Since we can't increase page size, let's calculate the efficiency and test concurrent requests
import concurrent.futures
import time

print("=== Plus API Pagination Analysis ===")
print(f"Products per page: 12")
print(f"Total products: 17,254")
print(f"Total pages needed: 1,438")
print(f"Time per request: ~1-2 seconds")
print(f"Estimated total time (sequential): {1438 * 1.5 / 60:.1f} minutes")

# Test concurrent requests to see if we can parallelize
def fetch_page(page_num):
    """Fetch a single page of products"""
    test_json = copy.deepcopy(json_data)
    test_json['screenData']['variables']['PageNumber'] = page_num
    
    try:
        response = requests.post(
            'https://www.plus.nl/screenservices/ECP_Composition_CW/ProductLists/PLP_Content/DataActionGetProductListAndCategoryInfo',
            cookies=cookies,
            headers=headers,
            json=test_json,
            timeout=10
        )
        
        if response.status_code == 200:
            data = response.json()
            return len(data['data']['ProductList']['List'])
        else:
            return 0
    except Exception as e:
        print(f"Error fetching page {page_num}: {e}")
        return 0

# Test fetching multiple pages concurrently
print("\n=== Testing Concurrent Requests ===")
start_time = time.time()

# Test with 5 different pages
test_pages = [1, 10, 50, 100, 200]
with concurrent.futures.ThreadPoolExecutor(max_workers=3) as executor:
    results = list(executor.map(fetch_page, test_pages))

end_time = time.time()
duration = end_time - start_time

print(f"Fetched {len(test_pages)} pages in {duration:.2f} seconds")
print(f"Results: {dict(zip(test_pages, results))}")
print(f"Average time per page: {duration/len(test_pages):.2f} seconds")

if all(r == 12 for r in results):
    print("\n✓ Concurrent requests work and return consistent results")
    concurrent_time = 1438 / 3 * (duration/len(test_pages)) / 60  # Assuming 3 workers
    print(f"Estimated time with 3 concurrent workers: {concurrent_time:.1f} minutes")
else:
    print("\n⚠ Inconsistent results with concurrent requests")
```
